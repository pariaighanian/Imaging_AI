### 1. Read a grayscale image (for example cameraman.tif ) with Matlab (imread()) and save it again in the formats GIF, TIF, PNG and JPEG (imwrite()). How does the file size differ between these formats?

```matlab
% Read the grayscale image
img = imread('cameraman.tif');

% Save the image in different formats
imwrite(img, 'cameraman.gif');
imwrite(img, 'cameraman.png');
imwrite(img, 'cameraman.tiff');  % Save as .tiff (TIF)
imwrite(img, 'cameraman.jpg');   % Save as .jpg (JPEG)

% Get file information and sizes
info_gif = dir('cameraman.gif');
info_png = dir('cameraman.png');
info_tiff = dir('cameraman.tiff');
info_jpg = dir('cameraman.jpg');

% Display file sizes
fprintf('File Size of GIF: %.2f KB\n', info_gif.bytes / 1024);
fprintf('File Size of PNG: %.2f KB\n', info_png.bytes / 1024);
fprintf('File Size of TIFF: %.2f KB\n', info_tiff.bytes / 1024);
fprintf('File Size of JPEG: %.2f KB\n', info_jpg.bytes / 1024);
```

### 2. What is the range of the intensity values of the image? Scale the gray value image to the value range [0, 1] and display the image both as a gray value and as a binary image on the monitor (imshow()).

```matlab
% Read the grayscale image
img = imread('cameraman.tif');

% Get the range of intensity values
min_intensity = min(img(:));  % Minimum intensity value
max_intensity = max(img(:));  % Maximum intensity value

% Display the intensity range
fprintf('Intensity Value Range: [%d, %d]\n', min_intensity, max_intensity);

% Scale the image to the range [0, 1]
img_scaled = double(img) / 255;  % Convert to double and scale to [0, 1]

% Display the scaled grayscale image
figure;
imshow(img_scaled);
title('Grayscale Image Scaled to [0, 1]');

% Convert the image to binary
threshold = 0.5;  % Choose a threshold value (you can change this as needed)
img_binary = imbinarize(img_scaled, threshold);  % Convert to binary using the threshold

% Display the binary image
figure;
imshow(img_binary);
title('Binary Image');
```
